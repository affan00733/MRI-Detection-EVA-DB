```python
cursor.execute("""DROP UDF TorchVissionPretrainedModel;""")
response = cursor.fetch_all()
print(response)

cursor.execute("""CREATE UDF IF NOT EXISTS 
                  TorchVissionPretrainedModel
                  INPUT (frame NDARRAY UINT8(3, ANYDIM, ANYDIM)) 
                  OUTPUT (results NDARRAY FLOAT32(ANYDIM, ANYDIM)) 
                  TYPE  Classification 
                  IMPL 'TorchVissionPretrainedModel.py';
        """)
response = cursor.fetch_all()
print(response)
```
